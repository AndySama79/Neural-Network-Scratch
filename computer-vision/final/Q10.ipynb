```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sep_conv(img, kernel_hori, kernel_vert):
    conv_hori = cv2.filter2D(img, -1, kernel_hori)
    conv_vert = cv2.filter2D(conv_hori, -1, kernel_vert)

    return conv_vert
```

```python
img = cv2.imread('images/building_1.jpg')
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
kernel_vert = np.array([1, 2, 1])
kernel_hori = np.array([[1],
                        [0],
                        [-1]])
```

```python
res = sep_conv(gray_img, kernel_hori, kernel_vert)
```

```python
plt.imshow(res, cmap='gray')
plt.axis('off')
plt.show()
```
